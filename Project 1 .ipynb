## Deleting repeated apps

As it was specified above, we want to have the highest number of apps reviews therefore we will make a dictionary that has the following features. 
1. Each dictionary key will be a unique app name. 
2. The value of each key will be the maximum from all reviews available from the original list.


```python
reviews_max={}
for app in android:
    name=app[0]
    n_reviews=float(app[3])
    if name in reviews_max  and reviews_max[name]< n_reviews:
        reviews_max[name]=n_reviews
    elif name not in reviews_max:
        reviews_max[name]=n_reviews        
```

*We want to make sure that our dictionary is correct, to check that out notice that*

```python
print("The expected length of the dictionary is:", len(android)-len(duplicate_apps))
print("The actual length of the dictionary is:", len(reviews_max))
```

**Therefore the dictionary is correct**

# Using the created dictionary to delete repeated apps. 

Remember that our criteria, to delete repeated apps was to select those apps with the highest number of reviews, but **some of the repeated apps have the same number of reviews**, therefore we need to be careful to avoid repeating apps again. Here is an example of an app, whose name is "Box" and has exactly the same number of reviews every time is repeated.

```python
for app in android: 
    name=app[0]
    if name=="Box":
        print(app)
        print("\n")
```

Taking into account what I just discussed above regarding the repeating apps we will write a code that does the following: 
* Initially, two lists named **android_clean**  and  **android_added** will be created. 
* After that, we will loop through the **android**  list.
  - At each iteration the name and the number of reviews will be assigned each to a variable 
  - If the aready assigned variable matches the number of reviews from our dictionary **AND**  the name does not belong to the already added list. The entire list will be appended to the **android_clean** list, and the name of the app will be appended to the **already_added** list. 
  - Notice that the condition "the name does not belong to the already added list" guarantee that we will not repeat an app if it has the same number of reviews each time it repeats. 



```python
android_clean=[]
already_added=[]
for app in android:
    name=app[0]
    n_reviews=float(app[3])
    if ( n_reviews == reviews_max[name] ) and ( name not in already_added ):
        android_clean.append(app)
        already_added.append(name)
```

In fact we see that there are many non_English apps, so we will create a function that returns false when the input is an string that has a character with al least one non-English character. 

```python
def is_english(x):
    for character in x:
        if ord(character)>127:
            return False
    return True
    
print(is_english('Instagram'))      
print(is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

The goal of this section is to remove the non_Eglish apps from our list, the problem is that the codes we have written so far do not take into account that there could be English apps that have symbols that fall out of ASCII range. In the following example we see an English app that is filtered for our function as non-English. 

```python
print(is_english('Bacterial Vaginosis 🇺🇸'))
```

Therefore we will modify our function a little bit and classify apps with more than three characters falling outside of the ASCII range as non_English apps and that will prevent us from loosing big amounts of information. 

```python
def is_english(x):
    vector=[]
    for character in x:
        if ord(character)>127:
            vector.append(character)
    if len(vector)>3:
        return False 
    else:
        return True        
    
print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('Instachat 😜'))
print(is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

Our function is not perfect but it will help us better filter English apps from non-English apps. We will now use the function we just built to create two new data sets that contain English apps only. 

```python
android_english=[]
ios_english=[]
for app in android_clean:
    if is_english(app[0]):
        android_english.append(app)

for app in ios:
    if is_english(app[2]):
        ios_english.append(app)

explore_data(android_english,0,3,True)
print("\n")
explore_data(ios_english,0,3,True)
```

# Frequency tables of genres 

At our company we follow a validation strategy that is composed of the following steps when a new application is released. 

1. A first version of the app is released for Android. 
2. If the app has a good response from users, we develop it further.
3. If the app is profitable after 6 months, we develop and iOS version. 

Taking into account that our goal is to develop our app for both markets we want to understand what are succesful profiles of apps in both markets, that is the reason because a frequency table for the `prime_genre` from the iOS data set, and a frequency table of the `Genre`  and `Category` from the android data set will be built. 

The ideal object to be created should be a frequency table and for our analysis we wil have to sort the results using the built-in `sorted()` function but the function does not work very well with a dictionary because is going to try to sort the keys of dictionary instead of sorting the values of the dictionary and the function would work better with tuples as we can see on the following examples. 



```python
a_list=[50,20,100]
print(sorted(a_list))
print('\n')
print(sorted(a_list, reverse = True))
```

As it is clear from the example above the `sorted()` function works very well with a list, now, lets see that it does not work very well with a dictionary. 

```python
a_dictionary={'Genre_1':50,'Genre_2':20,'Genre_3':100}
print(sorted(a_dictionary))
print('\n')
print(sorted(a_dictionary,reverse = True))
```

The result we were expected to have was `['Genre_2','Genre_1','Genre_3']`  for the application of the `sorted()` function, and similarly the result we got for `reverse =  True` was not correct either. 

Now, notice that the `sorted()` function does work very well with tuples, as you can see on the following example. 

```python
a_tuple = [(50,'Genre_1'),(20,'Genre_2'),(100,'Genre_3')]
print(sorted(a_tuple))
print('\n')
print(sorted(a_tuple, reverse = True))
```

Therefore, we are going to write code that has the following features.  
1. Initially, given a data set which is expected to be a list of lists, and an index, a frequency table will be created. 
2. Once the frequence table has been created a function that converts the dictionary into tuples will use the `sorted()` function to print the final result. 

```python
def freq_table(dataset,index):
    dictionary={}
    for entry in dataset:
        if entry[index] not in dictionary:
            dictionary[entry[index]]=1
        else:
            dictionary[entry[index]]+=1
    for key in dictionary:
        dictionary[key]=(dictionary[key]/len(dataset))*100
    return dictionary    



def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

We will now use the two functions just created to display the frequency tables of `prime_genre`, `Genre` and `Category`.

# Most popular apps by genre on the AppStore

One way to know what are the most popular apps in the AppstStore is by looking at the number of installs, but that information is missing from the the AppleSore data set therefore we will analyze the information, using the the number of reviews per genre.  Which we do down below. 

```python
genres_ios = freq_table(ios_final, -5)

for genre in genres_ios:
    total = 0
    len_genre = 0
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[6])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)
```

The highest amount of reviews comes from the "Navigation" genre, but that is highly influenced by GoogleMaps and Waze.
